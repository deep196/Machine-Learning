```python
#area & price 
area=  np.array([700,750,800,900,1000])
price= np.array([20,30,33,45,80])
```

```python
plt.figure()
plt.scatter(area,price)
plt.show()
```

```python
#random slope & intercept
slope=0.5
intercept=10
predicted_price=area*slope+intercept
```

```python
plt.figure()
plt.scatter(area,predicted_price)
plt.plot(area,predicted_price)
plt.show()
```

```python
#ordinary least square
area_mean=np.mean(area)
price_mean=np.mean(price)
```

```python
slope=np.sum((area-area_mean)*(price-price_mean))/np.sum((area-area_mean)**2)
```

```python
intercept=price_mean-(slope*area_mean)
```

```python
slope
```

```python
intercept
```

```python
predicted_price=area*slope+intercept
```

```python
plt.figure()
plt.scatter(area,price)
plt.plot(area,predicted_price)
plt.show()
```

```python
#linear
def poly_1(x):
    y=4*x+5
    return y

#Quadratic
def poly_2(x):
    y=6*(x**2)+4*x+5
    return y

#cubic
def poly_3(x):
    y=8*(x**3)+6*(x**2)+4*x+5
    return y

#plot 
def plot(x,y):
    plt.figure()
    plt.plot(x,y)
    plt.show()
```

```python
x=np.arange(-100,100)
```

```python
y=poly_1(x)
plot(x,y)
```

```python
y=poly_2(x)
plot(x,y)
```

```python
y=poly_3(x)
plot(x,y)
```
