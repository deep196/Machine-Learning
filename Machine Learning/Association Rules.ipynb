```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("GroceryStoreDataSet.csv", names = ['products'], sep = ',')
df.head()
```

```python
data = list(df["products"].apply(lambda x:x.split(",") ))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
encoded_data = te.fit_transform(data)
df = pd.DataFrame(encoded_data,columns=te.columns_)
df = df.replace(False,0)
df = df.replace(True,1)
df.head()
```

```python
df.shape
```

```python
scores = apriori(df, min_support = 0.2, use_colnames = True)
scores
```

```python
scores_ar = association_rules(scores, metric = "confidence", min_threshold = 0.6)
scores_ar
```
